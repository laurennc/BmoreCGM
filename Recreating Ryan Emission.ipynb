```python
import yt
import numpy as np
import matplotlib.pyplot as plt
from emission_functions import *
```

```python
fn = '/Users/dalek/data/Ryan/r0058_l10/redshift0058'
ds = yt.load(fn, file_style="%s.grid.cpu%%04i")
pos2 = [ 0.40320258,  0.4718025 ,  0.46123817]
rad = 316.8648 ##kpc

data =ds.sphere(pos2,(rad,'kpc'))
```

```python
import brewer2mpl as brew

bmap = brew.get_map('PuRd','Sequential',9)
cmap = bmap.get_mpl_colormap(N=1000, gamma=2.0)
```

```python
proj = yt.ProjectionPlot(ds,'x',('gas','Emission_CIII_977'),center=pos2,width=(320.,'kpc'),
                         data_source=data)
proj.set_cmap(('gas','Emission_CIII_977'),cmap)
proj.set_zlim(('gas','Emission_CIII_977'),1e-5,1e5)
proj.show()
```
