```python
data = np.genfromtxt('positions.csv',delimiter=',')
x_range = (data[:,0].min(),data[:,0].max())
y_range = (data[:,1].min(),data[:,1].max())
```

```python
df = pd.read_csv('positions.csv',names=['x','y','z'])
```

```python
plot_width = 400
plot_height = 400
x_range = (0.,1.)
y_range = (0.,1.)
```

```python
def base_plot(tools='pan,wheel_zoom,reset',plot_width=plot_width,plot_height=plot_height,**plot_args):
    p = figure(tools=tools,plot_width=plot_width,plot_height=plot_height,x_range=x_range,y_range=y_range,
            outline_line_color=None, min_border=0, min_border_left=0,
            min_border_right=0,min_border_top=0,min_border_bottom=0,**plot_args)
    p.axis.visible = False
    p.xgrid.grid_line_color = None
    p.ygrid.grid_line_color = None
    return p
```

```python
def create_image(x_range,y_range,w=plot_width,h=plot_height):
    cvs = datashader.Canvas(plot_width=w,plot_height=h,x_range=x_range, y_range=y_range)
    agg = cvs.points(df,'x','y')#,datashader.count('x'))
    img = tf.shade(agg,cmap=['lightblue','darkblue'],how='log')
    return img
```

```python
p = base_plot()
```

```python
InteractiveImage(p, create_image)
```
